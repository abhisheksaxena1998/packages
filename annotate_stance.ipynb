```
import pandas as pd
import numpy as np
import requests

import matplotlib.pyplot as plt

df=pd.read_csv('stephentwigg.csv')
df
```

```
url="http://104.248.174.180/api/stance/brexit_tweets/"

stance=[]
for i in df['tweet']:
    def post_json(url,json_data):

      res= requests.post(url,json=json_data)
      #print (res.json())
      jsonobj=res.json()
      stance.append(jsonobj[0]['Stance'])
      print (jsonobj[0]['Stance'])
    post_json(url,{"Tweets":[i]})
```

```
df['date']=df['date'].apply(lambda x :x[5:7])
df['stance']=stance
df.to_csv('stephentwigg_stance.csv')
df=pd.read_csv('stephentwigg_stance.csv')
df
```
