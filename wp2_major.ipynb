```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

dff=pd.read_csv('output.csv')
dff
```

```python
d11=pd.read_csv("output1.csv")
d11
```

```python
d2=d11[d11['month']==5]
d2
len(d2)
```

```python
b=d2.groupby('stance',sort=True).count()
b
```

```python
x=[]

for i in b.index:
    x.append(i)
x
```

```python
y=[]

for i in b['Unnamed: 0']:
    y.append(i)
y
```

```python
fig, ax = plt.subplots(figsize=(15,7))
plt.bar(x, y, edgecolor = 'black', linewidth=2)

ax.set_xlabel('Stance of post',fontsize=20)
ax.set_ylabel('Frequency',fontsize=20)
ax.set_title("Stance of post and their frequency for May",fontsize=20)

ax = plt.gca()
ax.tick_params(axis = 'both', which = 'major', labelsize = 15)
```
