```
from hatesonar import Sonar
sonar = Sonar()
s=sonar.ping(text="At least I'm not a nigger")
a=s['classes'][0]['confidence']
b=s['classes'][1]['confidence']
c=s['classes'][2]['confidence']
maxi=max(a,b,c)
if a==maxi:
    cos=s['classes'][0]['class_name']
elif b==maxi:
    cos=s['classes'][1]['class_name']
else:
    cos=s['classes'][2]['class_name']
cos          
```

```
import pandas as pd
import numpy as np

df=pd.read_csv("Mark1.csv")
df.replace('N/A',np.NaN)
#df=df.sample(frac=0.1,random_state=42)

len(df)
```

```
CoS=[]
ct=0
for i in df['text']:
    sonar = Sonar()
    s=sonar.ping(text=i)
    a=s['classes'][0]['confidence']
    b=s['classes'][1]['confidence']
    c=s['classes'][2]['confidence']
    maxi=max(a,b,c)
    if a==maxi:
        cos=s['classes'][0]['class_name']
    elif b==maxi:
        cos=s['classes'][1]['class_name']
    else:
        cos=s['classes'][2]['class_name']
    print (cos)
    ct+=1
    print (ct)    
    CoS.append(cos)   
```

```
df['classofspeech']=CoS
df.to_csv("Mark1ann.csv")
df
```
